```python
import matplotlib.pyplot as plt

df.boxplot(column="loan_int_rate", by="loan_status", grid=False)
plt.title("Loan Interest Rate by Loan Status")
plt.suptitle("")
plt.xlabel("Loan Status (0 = No Default, 1 = Default)")
plt.ylabel("Interest Rate (%)")
plt.show()
```

```python
df.boxplot(column="loan_percent_income", by="loan_status", grid=False)
plt.title("Loan Percent of Income by Loan Status")
plt.suptitle("")
plt.xlabel("Loan Status (0 = No Default, 1 = Default)")
plt.ylabel("Loan Percent of Income")
plt.show()
```
